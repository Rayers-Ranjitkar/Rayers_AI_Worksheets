# **2. Exploratory Data Analysis Practice Exercise-1**

```python
old_null_count = titanic_df['Age'].isnull().sum()
print(f"Total old_data_null_Count = {old_null_count}")

#Handling Null data by filling them with median...............
median_age = titanic_df['Age'].median()
#print(median_age)

titanic_df['Age'].fillna(median_age, inplace=True) #fillna fills the cols data where the data is null and returns a new df but, When inplace=True: The original DataFrame is modified directly. The method returns None, not a new DataFrame. Many pandas methods like .fillna(), .drop(), .rename(), .sort_values() have the parameter inplace.
print(f"Total new_data_null_Count = {titanic_df['Age'].isnull().sum()}")
```

**Which passenger had the highest fare paid relative to their age?**

```python
# 1. Add a column to the Titanic dataset called 'fare_per_year', containing the Fare divided by the Age of the passenger (Fare / Age)

titanic_df['fare_per_year'] = titanic_df['Fare']/titanic_df['Age'] #here we had to deal with each row data operation on adjacent each row data so normal works. Transform becomes complex here -> titanic_df['fare_per_year'] = titanic_df[['Fare', 'Age']].transform(lambda x: x['Fare'] / x['Age']) # .transform() is mainly used after groupby(). #It applies a function to each group and returns a Series with the same index as the original DataFrame.#This is crucial when you want to perform group-wise operations but keep the original DataFrame shape. Where it makes more sense->  titanic_df['avg_fare_per_class'] = titanic_df.groupby('Pclass')['Fare'].transform('mean') -> When we have a grouped operation, like “average Fare per Pclass for each passenger”: -> first we need to group them and calculate avg of each group right -> Here .transform() is necessary because you want the result to have the same number of rows as the original DataFrame.
titanic_df.head()
#titanic_df.drop('fare_per_age', axis=1, inplace= True)
```

```python
# 2. Subset rows where 'fare_per_year' is higher than 5, assigning this to high_fare_age. (subset = select)

high_fare_age = titanic_df[titanic_df['fare_per_year'] > 5]
high_fare_age.head()
#print(high_fare_age)

```

```python
# 3. Sort high_fare_age by descending 'fare_per_year', assigning this to high_fare_age_srt

high_fare_age_sorted = high_fare_age.sort_values(by="fare_per_year",ascending = False) #sort_data -> sort_values, sort_index (to sort the index)
high_fare_age_sorted.head()
```

```python
#4. # Select only the 'Name' and 'fare_per_year' columns of high_fare_age_srt and save the result as 'result'
result_df = high_fare_age_sorted[['Name','fare_per_year']]
print(result_df)
```

# **3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:**




**What percent of the total fare revenue came from each passenger class?**

```python
#1. Calculate the total Fare paid across all passengers in the Titanic dataset.

overall_total_fare_paid = titanic_df['Fare'].sum()
print(f"The total fare paid by all pessengers is = {overall_total_fare_paid}")
```

```python
#2. Subset for passengers in first class (Pclass is 1) and calculate their total fare.

#titanic_df.groupby('Pclass').filter(lambda x: x==1) #group_by is used only where we have say same categories data say A and his purchases in the rows and we want to find unique A only with all his purchases totaled

first_class_psg = titanic_df[titanic_df["Pclass"] == 1]
#first_class_psg.head()
total_fare_first_class = first_class_psg['Fare'].sum()
print(total_fare_first_class)

```

```python
#3. Do the same for second class (Pclass is 2) and third class (Pclass is 3).

second_class_psg= titanic_df[titanic_df["Pclass"] == 2]
second_class_psg.head()
total_fare_second_class = second_class_psg['Fare'].sum()
print(total_fare_second_class)

#for third class
third_class_psg= titanic_df[titanic_df["Pclass"] == 3]
third_class_psg.head()
total_fare_third_class = third_class_psg['Fare'].sum()
print(total_fare_third_class)
```

```python
#4. Combine the fare totals from first, second, and third classes into a list.


total_fare_list_py = [total_fare_first_class, total_fare_second_class, total_fare_third_class] #python_list
#total_fare_list_np = np.arrays(total_fare_list_py) #numpy_arr_to_do_division
total_fare_series_pd = pd.Series(total_fare_list_py,index=[1,2,3] ) #pandas_list_aka_series(cols) #index = [1,2,3] -> start index from 1

# print(total_fare_series_pd)

#5. Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.
overall_total_fare = (total_fare_series_pd / overall_total_fare_paid)*100
print(f"Overall total fare : \n{overall_total_fare}")
print()
print("-"*100)

#Ans: Fare from 1st class was 63.34% of the total fare collected. Similarly, 2nd class was 13.24% and 3rd class was 23.40%.

for key,values in overall_total_fare.items(): #we can iterate pandas series like a dictionary aka js object
  print(f"Pessenger Class {key} = {values:.2f}%")

```

**What percent of the total number of passengers on the Titanic belonged to each age group (e.g., child, adult, senior)?**

```python
#1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age 18<64), and "senior" (age 65 and above).

titanic_df['age_group'] = titanic_df['Age'].apply(lambda x: 'child' if x<18 else('adult' if x<64 else 'senior')) #.apply -> It applies a function to each element of a Series (column) or to each row/column of a DataFrame.
#Can also do it by pd_cut()
titanic_df.head(20)
```

```python
#2. Calculate the total number of passengers on the Titanic.

total_passengers = titanic_df['PassengerId'].count()
print(total_passengers)
```

```python
# 3. Count the number of passengers in each age group.

passenger_count_by_age_group = titanic_df.groupby('age_group')['PassengerId'].count() #titanic_df.groupby('age_group') -> Groups the DataFrame by the values in the 'Category' column ('A' and 'B'). followed by value -> ['Value'] Selects the column to perform aggregation on. and .counts() finally counts the total data in the cols
print(passenger_count_by_age_group)
```

```python
#4. Divide the count of each age group by the total number of passengers to get the proportion of passenger in each age group.

proportion_by_age_group = (passenger_count_by_age_group/total_passengers)*100 ##passenger_count (by_age_group) gives counts per group, not individual ages #hmmm then where did the age_group go ? -> groupby('age_group') → splits the DataFrame into groups based on the age_group column. So internally, pandas now knows which rows belong to 'Adult', 'Child', 'Senior'. #counts = passenger_count_by_age_group.values  # [765, 113, 13] #we can do .values to use them separately when in need
print(proportion_by_age_group)
print()
print("-"*100)

for key,values in proportion_by_age_group.items(): #we can iterate pandas series like a dictionary aka js object
  print(f"{key} = {values:.2f}%")
```

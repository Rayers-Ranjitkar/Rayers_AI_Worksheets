# **3.1. Problem - 1A - Single Year HDI Exploration (Latest Year: 2022)**

**1. Extracting latest year**

```python
import pandas as pd

hdi_df= pd.read_csv("/content/drive/MyDrive/Dataset/Human_Development_Index_Dataset.csv",encoding="cp1252"); #As the data throws utf-8 error we need to make it's encoding either cp1252 or latin1
hdi_df.head()

# Identifing unique years in the dataset
print("Unique years in the dataset:")
print(hdi_df["year"].unique())

# Filtering the dataset for year 2022
hdi_2022_df = hdi_df[hdi_df["year"] == 2022]
```

**2. Data Exploration**

```python
print("Filtered dataset for year 2022:")
print("-" * 60)

# Displaying the first 10 rows of the 2022 dataset.
hdi_2022_df.head(10)
```

```python
# Counting number of rows and columns
rows, columns = hdi_2022_df.shape #Tuple(206x30) unpacking
print("Number of rows:", rows)
print("Number of columns:", columns)

print("-" * 60)

# Printing Column names and their data types
print("Column names and data types:")
print()
print(hdi_2022_df.dtypes)

print("Column names and their data types:")
print("-" * 60)
```

**3. Checking missing values and cleaning the data**

```python
# Checking missing values in each column (2022 dataset)
print("Missing values in each column (Year 2022):")
print()

missing_values = hdi_2022_df.isnull().sum()
print(missing_values)
```

```python
# Inspecting dataset
print("Checking data types if numeric column is stored as text:")
print()
print(hdi_2022_df.dtypes)
print("-" * 60)
# Finding: No numeric columns stored as text was found
```

```python
# Inspecting dublicate rows:
dup_count = hdi_2022_df.duplicated().sum()
print("Number of duplicate rows:", dup_count)

# Inspecting special characters (e.g., “–”) representing missing data.
special_char_count = (hdi_2022_df == "–").sum().sum() #first sum() counts per column and second sum() adds everything to get one number
print("Total count of '-' values in dataset:", special_char_count)
```

```python
# Applying necessary cleaning steps:
# No data types needs to be converted and no removing duplicates required as there are no duplicated rows

# ---------------------------------------------------
# Handling Missing Values:

# First, we checked the number of missing values in each column.
# If only a very small number of values are missing, we drop those rows
# because the data loss is negligible and it avoids making assumptions.

# Then,
# We calculated IQR to check whether outliers exist in numeric columns.
# If outliers exist, median is used since it is less affected by extreme values.
# If outliers do not exist, mean is used.

for each_col in hdi_2022_df.columns:

    # Checking missing values
    null_count = hdi_2022_df[each_col].isnull().sum()

    # Proceeding only if missing values exist
    if null_count > 0:

        # Handling only numeric columns
        if hdi_2022_df[each_col].dtype != 'object':

            print("Column:", each_col)
            print("Missing values:", null_count)

            # If very few missing values, dropping rows
            if null_count <= 4:
                print("Very few missing values → Dropped these null rows")
                hdi_2022_df = hdi_2022_df.dropna(subset=[each_col])

            # Else using IQR-based imputation
            else:
                print("Using IQR to decide mean or median:")

                col_values = hdi_2022_df[each_col].dropna() # removing NaNs temporarily to calculate Q1, Q3 and IQR and filling it later and storing it in temporary df col_values

                Q1 = col_values.quantile(0.25)
                Q3 = col_values.quantile(0.75)
                IQR = Q3 - Q1

                lower_bound = Q1 - 1.5 * IQR
                upper_bound = Q3 + 1.5 * IQR

                outliers = col_values[ (col_values < lower_bound) | (col_values > upper_bound)]

                if len(outliers) > 0:
                    print("Outliers found → Filling with MEDIAN")
                    hdi_2022_df[each_col] = hdi_2022_df[each_col].fillna(col_values.median()) #filling original df
                else:
                    print("No outliers found → Filling with MEAN")
                    hdi_2022_df[each_col] = hdi_2022_df[each_col].fillna(col_values.mean())

            print()
```

```python
print("Missing values after handling:",hdi_2022_df.isnull().sum().sum())
```

**4. Basic Statistics**

```python
# Creating country-only dataset (excluding regions world, and development groups)
country_only_df_2022 = hdi_2022_df[hdi_2022_df["iso3"].str.len() == 3].copy()
```

```python

# Calculating Mean, Median and Standard Deviation of HDI
mean_hdi = country_only_df_2022["hdi"].mean()
median_hdi = country_only_df_2022["hdi"].median()
std_hdi = country_only_df_2022["hdi"].std()

print("Mean HDI (2022):", round(mean_hdi,3) )
print("Median HDI (2022):", round(median_hdi,3) )
print("Standard Deviation of HDI (2022):", round(std_hdi,3) )

print("-" * 60)

# Country with highest HDI
max_hdi = country_only_df_2022["hdi"].max()
highest_hdi_country = country_only_df_2022[country_only_df_2022["hdi"] == max_hdi]

print("Country with highest HDI in 2022:")
print(highest_hdi_country[["country", "hdi"]])

print("-" * 60)

# Country with lowest HDI
min_hdi = country_only_df_2022["hdi"].min()
lowest_hdi_country = country_only_df_2022[country_only_df_2022["hdi"] == min_hdi]

print("Country with lowest HDI in 2022:")
print(lowest_hdi_country[["country", "hdi"]])
```

**5. Filtering and Sorting**

```python
# Filtering countries with HDI greater than 0.800
high_hdi_countries = country_only_df_2022[country_only_df_2022["hdi"] > 0.800]

print("Countries with HDI greater than 0.800:")
print("Total countries:", len(high_hdi_countries))
print("-" * 60)

# 2) Sorting by Gross National Income (GNI) per Capita in descending order
sorted_high_hdi_countries = high_hdi_countries.sort_values(
    by="gross_inc_percap",
    ascending=False
)

# 3) Displaying top 10 countries
print("Top 10 countries (HDI > 0.800) sorted by GNI per Capita:")
print()
print(sorted_high_hdi_countries[["country", "hdi", "gross_inc_percap"]].head(10))
```

**6. Adding HDI Category Column:**

```python
# Creating a new column with default value
country_only_df_2022["HDI Category"] = "Low human development"

# Applying UNDP thresholds and updating the default column value
country_only_df_2022.loc[(country_only_df_2022["hdi"] >= 0.550) & (country_only_df_2022["hdi"] <= 0.699), "HDI Category"] = "Medium human development" #df.loc[ row , column ] = value
country_only_df_2022.loc[(country_only_df_2022["hdi"] >= 0.700) & (country_only_df_2022["hdi"] <= 0.799), "HDI Category"] = "High human development"
country_only_df_2022.loc[country_only_df_2022["hdi"] >= 0.800, "HDI Category"] = "Very high human development"

print("Verifying:")
print()
print(country_only_df_2022[["country", "hdi", "HDI Category"]].head(10))
# Result:
# Updated dataframe includes new category column
# Countries are classified correctly based on HDI

country_only_df_2022.to_csv("HDI_category_added.csv", index=False)
print()
print("File saved as: HDI_category_added.csv")
```

**B. Generate Visualizations:**

**- Bar Chart: Average HDI by Region (2020–2022)**

```python
# Grouping the dataset by Region and Year

# Selecting non-country rows i.e. (rows with country as regions + human development group)
non_country_df = hdi_2020_2022_df[
    hdi_2020_2022_df["iso3"].str.len() != 3
]

# Excluding World and Human Development group
exclude_entities = [
    "World",
    "Very high human development",
    "High human development",
    "Medium human development",
    "Low human development"
]

region_df = non_country_df[non_country_df["country"].isin(exclude_entities) == False]
#print(region_df[["country", "year", "hdi"]].head())

# Grouping by region (country column) and year, and computing mean HDI
avg_hdi_region_year = (
    region_df
    .groupby(["country", "year"])["hdi"]
    .mean()
    .reset_index() #without this "year" becomes part of index and not a column due to grouping
)

print(avg_hdi_region_year.head())

```

```python
import matplotlib.pyplot as plt

for year in [2020, 2021, 2022]:
    year_data = avg_hdi_region_year[avg_hdi_region_year["year"] == year]

    plt.figure(figsize=(8, 4))
    plt.barh(year_data["country"], year_data["hdi"])

    plt.xlabel("Average HDI",labelpad=15)
    plt.ylabel("Region",labelpad=15)
    plt.title(f"Average HDI by Region ({year})",pad=15)

    plt.xlim(avg_hdi_region_year["hdi"].min() - 0.01,  #limiting the values to this range
    avg_hdi_region_year["hdi"].max() + 0.01)

    plt.show()
    print()
    print("-" * 80)
    print()
```

**– Box Plot: HDI Distribution for 2020, 2021, and 2022**

```python
# Filtering the dataset for the years 2020, 2021, and 2022 and dropping the 8 empty HDI rows.
hdi_2020_2022_df = hdi_df[hdi_df["year"].isin([2020, 2021, 2022])]
hdi_2020_2022_df = hdi_2020_2022_df.dropna(subset=["hdi"])


# Extracting HDI values separately for each year
hdi_2020 = hdi_2020_2022_df[hdi_2020_2022_df["year"] == 2020]["hdi"]
hdi_2021 = hdi_2020_2022_df[hdi_2020_2022_df["year"] == 2021]["hdi"]
hdi_2022 = hdi_2020_2022_df[hdi_2020_2022_df["year"] == 2022]["hdi"]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 4))

plt.boxplot(
    [hdi_2020, hdi_2021, hdi_2022],
    tick_labels=["2020", "2021", "2022"],
)

plt.xlabel("Year",labelpad=15)
plt.ylabel("HDI",labelpad=15)
plt.title("HDI Distribution for 2020, 2021, and 2022",pad=15)

plt.show()

# For Observation on the distribution differences:
print()
print("Median HDI values:")
print("2020:", hdi_2020.median())
print("2021:", hdi_2021.median())
print("2022:", hdi_2022.median())
```

The box plots show that the overall HDI distribution remains largely stable from 2020 to 2021.
There is a slight upward shift in the median HDI from 2021 to 2022, indicating gradual improvement in human development.   
The interquartile range (IQR) and overall spread remain similar across all three years, i.e. approximately between 0.6 and 0.8 across all three years
suggesting that the distribution in HDI levels across countries has not changed significantly, while improvements are incremental rather than drastic
which the slow and gradual nature of changes in human development.

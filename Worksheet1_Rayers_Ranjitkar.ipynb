```python
#1. Prove A.A−1 = I.

import numpy as np
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]]);

A_inv = np.linalg.inv(A)
multipliedResult = np.dot(A,A_inv) #It's same as A @ A_inv

print(np.round(multipliedResult) )
isEqual = np.array_equal(np.round(multipliedResult),np.eye(2,2))
print("Multiplied result is identity: ",isEqual)

```

```python
#2. Prove AB̸=BA

import numpy as np
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]]);

AB = A @ B
BA = B @ A
print("AB= \n",AB)
print("BA= \n",BA)

isEqual = np.array_equal(AB,BA)
print("Are AB and BA equal : ", isEqual)
```

```python
#3. Prove (AB)T = BT AT.

import numpy as np
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]]);

lhs = (A@B).T    #np.dot(A, B).T
rhs = (B.T @A.T) #np.dot(B.T, A.T)

print("LHS =\n", lhs)
print("RHS =\n", rhs)


isEqual = np.array_equal(lhs, rhs)
print("Are they equal? ->", isEqual)
```

**4.2.1. Element-wise Addition:**

```python
# Using Python Lists, perform element-wise addition of two lists of size 1,000,000.
# Measure and Print the time taken for this operation.

import random
import time

size = 1000000
list1 = []
list2 = []
result = []

#Start
start_time = time.time()

for i in range(size):
  random_number = random.random()
  list1.append(random_number)

for i in range(size):
  random_number = random.random()
  list2.append(random_number)

for i in range(size):
  result.append(list1[i] +list2[i])

#End
end_time = time.time()

print(result[:10:1])

print("Time taken for normal addition", round(end_time - start_time,2), "sec")

```

**4.2.3. Dot-product**

```python
#Using Python Lists, compute the dot product of two lists of size 1,000,000.
#Measure and Print the time taken for this operation.

import random
import time

size = 1000000
list1 = []
list2 = []
result = []

#Start
start_time = time.time()

for i in range(size):
  random_number = random.random()
  list1.append(random_number)

for i in range(size):
  random_number = random.random()
  list2.append(random_number)

dotProduct = 0
for i in range(size):
  dotProduct = dotProduct + (list1[i] *list2[i])

#End
end_time = time.time()

print("Dot-product = ",dotProduct)

print("Time taken for normal dot-product", round(end_time - start_time,2), "sec")
```

**4.2.3. Matrix-Multiplication**

```python
#Using Python lists, perform matrix multiplication of two matrices of size 1000x1000.
#Measure and print the time taken for this operation.

import random
import time

size = 1000
matrix1 = []
matrix2 = []
result = []

# Start
start_time = time.time()

# Generating two 1000x1000 matrix
for i in range(size):
    row = []
    for j in range(size):
        row.append(random.randint(0, 9))  # integers 0-9
    matrix1.append(row)

for i in range(size):
    row = []
    for j in range(size):
        row.append(random.randint(0, 9))
    matrix2.append(row)

# creating 1000x1000 result matrix
for i in range(size):
    row = []
    for j in range(size):
        row.append(0)
    result.append(row)

# performing matrix multiplication
for i in range(size): #track of current row
    for j in range(size): #track of current col
        for k in range(size): #to iterate over each element (makes each row x each col possible)
            result[i][j] += matrix1[i][k] * matrix2[k][j]

# End
end_time = time.time()

# Printing first result stored in (0,0)
print("Result[0][0] =", result[0][0])
print("Time taken for normal matrix multiplication:", round(end_time - start_time, 2), "sec")

```

```python
#Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this operation

import numpy as np
import time

size = 1000

matrix1 = np.random.randint(0, 10, (size, size)) #np.random.randint(lowValue,highValue,size)
matrix2 = np.random.randint(0, 10, (size, size))

#Start
start_time = time.time()

result = np.dot(matrix1, matrix2)   # or, result = matrix1 @ matrix2

end_time = time.time()

# Printing first result stored in (0,0)
print("Result[0][0] =", result[0, 0])
print("Time taken for NumPy matrix multiplication:", round(end_time - start_time, 2), "sec")

```
